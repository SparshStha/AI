# **Exercise 3**

Initially we will start with simple k-means algorithm implantation

````python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

random_state_value = 42
# Generate random data for clustering
np.random.seed(random_state_value)

data, _ = make_blobs(n_samples=300, centers=3, cluster_std=1.0, random_state=42)

# Function to perform k-means clustering and plot the results
def plot_kmeans_clusters(data, n_clusters):
    kmeans = KMeans(n_clusters=n_clusters, random_state=random_state_value)
    labels = kmeans.fit_predict(data)

    plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='viridis', edgecolor='k', s=50)

    plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', marker='x', s=200, label='Centroids')

    plt.title(f'K-means Clustering with {n_clusters} Clusters')
    plt.legend()

    plt.show()

plot_kmeans_clusters(data, n_clusters=2)
````

# **Exercise 4**

Read the script and understand how it works



This code demonstrates how to perform K-means clustering on a dataset and vusualize the result using Matplotlib. Here's a step-by-step explanation:

1. Importing Libraries

        import numpy as np
        import matplotlib.pyplot as plt
        from sklearn.datasets import make_blobs
        from sklearn.cluster import KMeans


* `numpy` is used for numerical operations.

* `matplotlib.pyplot` is used for plotting data.

* `sklearn.datasets` provides the `make_blobs` function to generate random data.

* `sklearn.cluster` provides the KMeans class for performing K-means clustering.

2. Setting a Random Seed

        random_state_value = 42
        np.random.seed(random_state_value)
* A random seed is set for reproducibility, ensuring that the random data generated and the clustering are the same each time the code is run.

3. Generating Random Data

        data, _ = make_blobs(n_samples=300, centers=3, cluster_std=1.0, random_state=42)

* `make_blobs` generates a dataset of 300 samples with 3 centers (clusters). The `cluster_std` parameter controls the standard deviation of the clusters. The `random_state` parameter ensures reproducibility

4. Defining the Function for K-means Clustering and plotting
        def plot_kmeans_clusters(data, n_clusters):
            kmeans = KMeans(n_clusters=n_clusters, random_state=random_state_value)
            labels = kmeans.fit_predict(data)

* A function `plot_kmeans_clusters` is defined, which takes the data and the number of clusters as arguments.
* `KMeans` is initialized with the specified number of clusters and the random state.
* The `fit_predict` method of `KMeans` is used to perform clustering on the data and return the cluster labels for each point.

5. Plotting the Data Points and Centroids

        plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='viridis', edgecolor='k', s=50)


* The data points are plotted using `plt.scatter`. The `c` parameter is set to the cluster labels, which colors the points according to their assigned cluster.
* `cmap='viridis'` specifies the color map to use.
`edgecolor='k'` sets the edge color of the points to black.
* `s=50` sets the size of the points.
        plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', marker='x', s=200, label='Centroids')

* The cluster centroids are plotted as red 'x' markers with a size of 200. These are the central points of each cluster.

6. Adding a Titel and Legend

        plt.title(f'K-means Clustering with {n_clusters} Clusters')
        plt.legend()


* A title is added to the plot, indicating the number of clusters.
* A legend is added to identify the centroids.

7. Displaying the Plot

        plt.show()
      
* The `plt.show()` function is called to display the plot.

8. Calling the Function
        plot_kmeans_clusters(data, n_clusters=2)

* The `plot_kmeans_clusters` function is called with the generated data and `n_clusters=2` to perform clustering and plot the results with 2 clusters.

# **Exercise 5**

Update the script to cluster the data with different cluster number ranging from 2 clusters to 10 (using loop).

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

random_state_value = 42
# Generate random data for clustering
np.random.seed(random_state_value)
data, _ = make_blobs(n_samples=300, centers=3, cluster_std=1.0, random_state=42)

# Function to perform k-means clustering and plot the results
def plot_kmeans_clusters(data, n_clusters):
    kmeans = KMeans(n_clusters=n_clusters, random_state=random_state_value)
    labels = kmeans.fit_predict(data)

    plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='viridis', edgecolor='k', s=50)
    plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', marker='x', s=200, label='Centroids')
    plt.title(f'K-means Clustering with {n_clusters} Clusters')
    plt.legend()
    plt.show()

# Loop to plot K-means clustering results for cluster numbers from 2 to 10
for n_clusters in range(2, 11):
    plot_kmeans_clusters(data, n_clusters)
````

# **Exercise 6**
Experiment with different random state value and comment on its effect on the algorithm.

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Function to generate random data for clustering
def generate_data(random_state_value):
    np.random.seed(random_state_value)
    data, _ = make_blobs(n_samples=300, centers=3, cluster_std=1.0, random_state=42)
    return data

# Function to perform K-means clustering and plot the results
def plot_kmeans_clusters(data, n_clusters, random_state_value):
    kmeans = KMeans(n_clusters=n_clusters, random_state=random_state_value)
    labels = kmeans.fit_predict(data)

    plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='viridis', edgecolor='k', s=50)
    plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', marker='x', s=200, label='Centroids')
    plt.title(f'K-means Clustering with {n_clusters} Clusters (Random State: {random_state_value})')
    plt.legend()
    plt.show()

# Generate data
data = generate_data(random_state_value=42)

# Experiment with different random state values
for random_state in [0, 21, 42, 84]:
    plot_kmeans_clusters(data, n_clusters=3, random_state_value=random_state)
````

# **Exercise 7**
Now we will use the iris dataset and use the k means algorithm to cluster it

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans

random_state_value = 42

# Load the Iris dataset
iris = load_iris()
data = iris.data

# Function to perform K-means clustering and plot the results
def plot_kmeans_clusters(data, n_clusters):
    kmeans = KMeans(n_clusters=n_clusters, random_state=random_state_value)
    labels = kmeans.fit_predict(data)

    plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='viridis', edgecolor='k', s=50)
    plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', marker='x', s=200, label='Centroids')
    plt.title(f'K-means Clustering with {n_clusters} Clusters')
    plt.legend()
    plt.show()

# Plot the data with different numbers of clusters
N = [2, 3, 4, 5, 6, 7]
for n in N:
    plot_kmeans_clusters(data, n_clusters=n)
````

**Data Generation**

````python
import numpy as np

from sklearn.datasets import make_blobs

n_samples = 1500
random_state = 170
transformation = [[0.60834549, -0.63667341], [-0.40887718, 0.85253229]]

X, y = make_blobs(n_samples=n_samples, random_state=random_state)
X_aniso = np.dot(X, transformation)  # Anisotropic blobs
X_varied, y_varied = make_blobs(
    n_samples=n_samples, cluster_std=[1.0, 2.5, 0.5], random_state=random_state
)  # Unequal variance
X_filtered = np.vstack(
    (X[y == 0][:500], X[y == 1][:100], X[y == 2][:10])
)  # Unevenly sized blobs
y_filtered = [0] * 500 + [1] * 100 + [2] * 10
````

Visualize the resulting data

````python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(12, 12))

axs[0, 0].scatter(X[:, 0], X[:, 1], c=y)
axs[0, 0].set_title("Mixture of Gaussian Blobs")

axs[0, 1].scatter(X_aniso[:, 0], X_aniso[:, 1], c=y)
axs[0, 1].set_title("Anisotropically Distributed Blobs")

axs[1, 0].scatter(X_varied[:, 0], X_varied[:, 1], c=y_varied)
axs[1, 0].set_title("Unequal Variance")

axs[1, 1].scatter(X_filtered[:, 0], X_filtered[:, 1], c=y_filtered)
axs[1, 1].set_title("Unevenly Sized Blobs")

plt.suptitle("Ground truth clusters").set_y(0.95)
plt.show()
````

Fit models and plot results

````python
from sklearn.cluster import KMeans

common_params = {
    "n_init": "auto",
    "random_state": random_state,
}

fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(12, 12))

y_pred = KMeans(n_clusters=2, **common_params).fit_predict(X)
axs[0, 0].scatter(X[:, 0], X[:, 1], c=y_pred)
axs[0, 0].set_title("Non-optimal Number of Clusters")

y_pred = KMeans(n_clusters=3, **common_params).fit_predict(X_aniso)
axs[0, 1].scatter(X_aniso[:, 0], X_aniso[:, 1], c=y_pred)
axs[0, 1].set_title("Anisotropically Distributed Blobs")

y_pred = KMeans(n_clusters=3, **common_params).fit_predict(X_varied)
axs[1, 0].scatter(X_varied[:, 0], X_varied[:, 1], c=y_pred)
axs[1, 0].set_title("Unequal Variance")

y_pred = KMeans(n_clusters=3, **common_params).fit_predict(X_filtered)
axs[1, 1].scatter(X_filtered[:, 0], X_filtered[:, 1], c=y_pred)
axs[1, 1].set_title("Unevenly Sized Blobs")

plt.suptitle("Unexpected KMeans clusters").set_y(0.95)
plt.show()
````

Possible solutions

````python
y_pred = KMeans(n_clusters=3, **common_params).fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=y_pred)
plt.title("Optimal Number of Clusters")
plt.show()
````

````python
from sklearn.mixture import GaussianMixture

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))

y_pred = GaussianMixture(n_components=3).fit_predict(X_aniso)
ax1.scatter(X_aniso[:, 0], X_aniso[:, 1], c=y_pred)
ax1.set_title("Anisotropically Distributed Blobs")

y_pred = GaussianMixture(n_components=3).fit_predict(X_varied)
ax2.scatter(X_varied[:, 0], X_varied[:, 1], c=y_pred)
ax2.set_title("Unequal Variance")

plt.suptitle("Gaussian mixture clusters").set_y(0.95)
plt.show()
````
